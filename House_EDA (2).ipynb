```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# reading the csv file
df = pd.read_csv('/content/raw_house_data - raw_house_data.csv')
```

```python
df
```

```python
# fetching the number of rows(orientations) and columns(features)
rows = df.shape[0]
cols = df.shape[1]
print(rows,cols)
```

```python
df.columns
```

```python
# fetching the data types of columns
df.dtypes
```

```python
# getting the statistics idea
df.describe()
```

```python
#checking for the missing values
print(df.isnull().sum())
```

```python
#Drop/removing  rows with missing values
clean_df = df.dropna()
```

```python
clean_df
```

```python
print(clean_df.duplicated().sum())
clean_df = clean_df.drop_duplicates()
clean_df.head(5)
```

```python
#Converting the datatypes with suitable ones i.e. bathrooms, garage, fireplaces, HOA
clean_df['bathrooms'] = clean_df['bathrooms'].astype('int')
clean_df['garage'] = clean_df['garage'].astype('int')
clean_df['fireplaces'] = clean_df['fireplaces'].astype('int')

clean_df.dtypes
```

```python
clean_df.plot(kind='scatter', x='sold_price', y='zipcode', alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Area which has largest carpet area or lowest carpet area i.e.e which goes sway from common house areas
fig, ax =plt.subplots(figsize=(8,5))
ax.scatter(clean_df['zipcode'],clean_df['sqrt_ft'] )
ax.set_xlabel('Zipcode')
ax.set_ylabel('Carpet Area')
```

```python
#Which houses are built
clean_df['year_built'].plot(kind='box', title='House built year')
plt.show()
```

```python
sns.set_style("whitegrid")

my_plot = sns.boxplot(x = 'year_built', y = 'sold_price', data = clean_df)
my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=90)

```

```python
clean_df.to_csv('raw-data-cleaned.csv')
```
